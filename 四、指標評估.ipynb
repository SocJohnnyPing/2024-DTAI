### 看看決策情況

```python
#轉為數值型態
df['OUTCOME'] = pd.to_numeric(df['OUTCOME'], errors='coerce').astype('Int64')
df['train'] = pd.to_numeric(df['train'], errors='coerce').astype('Int64')
```

```python
TP = ((df['OUTCOME'] == 1) & (df['train'] == 1)).sum()
TN = ((df['OUTCOME'] == 0) & (df['train'] == 0)).sum()
FP = ((df['OUTCOME'] == 0) & (df['train'] == 1)).sum()
FN = ((df['OUTCOME'] == 1) & (df['train'] == 0)).sum()

print(f"True Positives (TP): {TP}")
print(f"True Negatives (TN): {TN}")
print(f"False Positives (FP): {FP}")
print(f"False Negatives (FN): {FN}")

# 計算準確率
accuracy = (TP + TN) / len(df)
print(f"準確率: {accuracy:.2%}")

# 計算其他指標
precision = TP / (TP + FP) if (TP + FP) > 0 else 0
recall = TP / (TP + FN) if (TP + FN) > 0 else 0
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

print(f"Precision: {precision:.2%}")
print(f"Recall: {recall:.2%}")
print(f"F1 Score: {f1_score:.2%}")
```

在先前的討論中，我提到了在合併資料框時將某些欄位轉換為集合的方法。這是因為在某些情況下，將資料轉換為集合可以幫助我們處理資料。

在 Python 中，集合（Set）是一個無序且不重複的元素集合。當我們需要快速查找元素是否存在於集合中時，使用集合比列表更高效，因為集合使用了哈希表來存儲元素，這使得查找操作的時間複雜度為 O(1)。

在先前的例子中，當我們需要找出某個資料框中某一欄位的唯一值時，將該欄位轉換為集合可以幫助我們快速找出唯一值，同時也可以去除重複值。這在合併資料框時尤其有用，因為我們可能需要對某些欄位進行比較，查看它們是否存在於另一個資料框中。

總之，將資料轉換為集合的主要目的是為了快速查找唯一值和比較資料，以及去除重複值。這可以提高程式的效率並簡化操作。
